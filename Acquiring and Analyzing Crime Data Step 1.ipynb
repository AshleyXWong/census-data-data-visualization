```python
#there seems to be a lot of NaN's in the "DISCHARGED_DT" column
#this checks to see if the number of NaN's in the column is equal to the number of rows in the dataset
inmate_data['DISCHARGED_DT'].isnull().sum() == len(inmate_data.index)
inmate_data
```

```python
#now to check if the rest of the columns have missing data
#return the column names that have missing data and how many rows have missing data
lst = []
for column_name in inmate_data.columns:
    if inmate_data[column_name].isnull().sum() > 0:
        lst.append((column_name, inmate_data[column_name].isnull().sum()))
lst
```
